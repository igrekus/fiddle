```
import pandas as pd
import seaborn as sns

df = pd.read_csv('profile.csv')
df = df.sort_values('alias')

miss = df[df['fn'].str.startswith('miss')]
dupe = df[df['fn'].str.startswith('dupe')]
```

```
miss = df[df['fn'].str.startswith('miss')]
miss = miss.drop(miss[miss.alias == '07.py'].index)
sns.barplot(x='alias', y='best', data=miss)
```

```
miss = df[df['fn'].str.startswith('miss')]
sns.barplot(x='alias', y='avg', data=miss)
```

```
miss = df[df['fn'].str.startswith('miss')]
miss = miss.drop(miss[miss.alias == '04.py'].index)
miss = miss.drop(miss[miss.alias == '07.py'].index)
sns.barplot(x='alias', y='worst', data=miss)
```

```
dupe = df[df['fn'].str.startswith('dupe')]
dupe = dupe.drop(dupe[dupe.alias == '05.py'].index)
sns.barplot(x='alias', y='best', data=dupe)
```

```
dupe = df[df['fn'].str.startswith('dupe')]
dupe = dupe.drop(dupe[dupe.alias == '04.py'].index)
dupe = dupe.drop(dupe[dupe.alias == '05.py'].index)
dupe = dupe.drop(dupe[dupe.alias == '07.py'].index)
sns.barplot(x='alias', y='avg', data=dupe)
```

```
dupe = df[df['fn'].str.startswith('dupe')]
dupe = dupe.drop(dupe[dupe.alias == '04.py'].index)
dupe = dupe.drop(dupe[dupe.alias == '05.py'].index)
dupe = dupe.drop(dupe[dupe.alias == '07.py'].index)
sns.barplot(x='alias', y='worst', data=dupe)
```
